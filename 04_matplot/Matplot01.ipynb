## Plot 직선 그리기

```python
'''
1. plot()
::
직선 or 꺾은 선 형태의 그래프를 그릴 수 있다
즉, 선형그래프를 그릴 수 있다

** 선형|비선형

구조
plot([x],[y])
리스트 하나 : [y]... [x]는 자동으로 0부터 1씩 증가
'''
# 1) 간단한 데이터를 하나 만들어서...(x,y) 직선 혹은 꺾은 선 그려보자
# 꺾은 선
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()

# 2) 직선
plt.plot([10,20,30,40])
plt.show()
```

## Plot 그래프에 옵션추가하기

```python
'''
1-2. 그래프 옵션
::
1) 'o' : 지점을 점으로 표시
1-1) 'ro' : 점을 빨간색으로 표시... 응용가능
2) '--' : 점선
3) 's' : 지점을 사각형점으로 표시
4) '^' : 지점을 삼각형점으로 표시

5) .title(str) : 제목 달기
6) 'label' ~ .legend() : 라벨 달고... legend해야 보임
7) .xlabe(), .ylabel() : x, y축에 이름 달기
8) .grid() : 그리드 그리기

9) .figure() : 그래프 틀 지칭
- 원하는 사이즈의 피규어를 만들 수 있다
- plot 전에 써야 함
- 'figsize=' : 튜플형태 (w,h)

10) 'lw' : line width
- 기본이 3
'''
plt.plot([1,2,3,4],[1,4,9,16],'o')
plt.show()

plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.show()

plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.title('Graph Test')
plt.show()
```

```python
'''
방법 1) plot( ) 안에 여러 선 넣기

방법 2) plot을 여러 지정해주고 plt.show() 하기
'''
```

```python
# 0~5미만까지 0.2 간격으로 출력
t = np.arange(0,5,0.2)
t

plt.plot(t,t,'--')
plt.show()
```

```python
# plot 안에 여러 자료 넣으면 됨
plt.plot(t,t,'--',t,t**2,'gs',t,t**3,'^')
plt.show()
```

```python
# 라벨달기
plt.plot(t,t,'r--', label='tips')
plt.plot(t,t**2,'bs', label='bill')
plt.plot(t,t**3,'g^', label='layer')
plt.legend()
plt.title('Wrapped Graphs')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
plt.show()
```

## Pandas 데이터 이용한 Visualization

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# 1. 딕셔너리로 df 만들기
np.random.seed(100)
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
        'd':np.random.randn(50)}
data['b']= data['a']+10 * np.random.randn(50)
data['d']= np.abs(data['d']*100)
```

```python
data = DataFrame(data)
data.head()
```

### 산점도(Scatter) 그래프 그리기

```python
'''
scatter() 함수로 그린다
산점도는 x,y에 해당하는 데이터들의 상관관계를 표시할 때
데이터들이 어떻게 흩어져 있는가를 알기 위해 사용

2개의 축을 기준으로 데이터가 얼마나 퍼져있는지(분포도)를 알 수 있기 때문에
산포도 혹은 산점도로 불린다.
'''
```

```python
'''
scatter() 속성
1) s = strong, marker size
2) c = color 관련
---- 많은 양의 데이터를 다룰 때 위 속성들을 잘 쓰지 않는다.
---- cmap을 주로 씀
3 marker
4) cmap : 템플릿

'''
# plt.scatter([1,2,3,4],[10,20,30,40], s=[100,200,300,400],
#            c=['yellow','green','magenta','red'])
plt.scatter([1,2,3,4],[10,20,30,40], s=[100,200,300,400],
            c=range(4),
            cmap='jet')
plt.colorbar()
plt.show()
```

```python
# plt.scatter()
# data를 지정해주면 컬럼명만 가져오면 됨
plt.scatter('a','b',data=data) 
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

```python
# 특정 지점을 기준으로 색 나눈기
import seaborn as sns

cuthalf = (data['a']>25)&(data['b']>30)
data['color'] = np.where(cuthalf==True,'red','blue')
sns.regplot(x=data['a'], y=data['b'],scatter_kws={'facecolors':data['color']},fit_reg=False)
plt.title('Scatter Plot with Colors', fontsize=20)
plt.show()
```

## Subplot

### subplot으로 plot(), scatter(), bar() 동시에 사용하기

```python
names = ['group_A','group_B','group_C'] # x축으로 사용
values = [1,10,100] # y축으로 사용
```

```python
# 인자값 1 = figure 영역이 첫 번째라는 의미
plt.figure(1, figsize=(9,3))

# 1= 첫 번째 영역, 3= 총 3개, 1 = 3개 중 첫 번째 차트
plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
plt.show()
```
