### Pearson Correlation Coefficient

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” Collaborative Filteringì—ì„œ ìœ ì‚¬ë„ ì¸¡ì • ë©”íŠ¸ë¦­ìœ¼ë¡œ ì£¼ë¡œ ì‚¬ìš©ë˜ë©°, ë‘ ë²¡í„°ì˜ ì„ í˜• ìƒê´€ê´€ê³„ë¥¼ í™•ì¸í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë©ë‹ˆë‹¤. ì‹ìœ¼ë¡œ í‘œí˜„í•˜ìë©´ ì•„ë˜ì™€ ê°™ìŠµë‹ˆë‹¤.

$$
similarity(\textbf x,\textbf y)=r_{\textbf x\textbf y} = \frac{\sum (x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum (x_i - \bar x)^2}\sqrt{\sum (y_i - \bar y)^2}}
$$

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” +1ê³¼ -1 ì‚¬ì´ì˜ ê°’ì„ ê°–ìŠµë‹ˆë‹¤. +1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ì–‘ì˜ ì„ í˜• ìƒê´€ê´€ê³„ê°€ ìˆë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•˜ê³  -1ì— ê°€ê¹Œìš¸ìˆ˜ë¡ ê°•í•œ ìŒì˜ ì„ í˜• ìƒê´€ê´€ê³„ê°€ ìˆë‹¤ëŠ” ê²ƒì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ë˜í•œ 0ì€ ì•„ë¬´ëŸ° ì„ í˜• ìƒê´€ê´€ê³„ê°€ ì—†ìŒì„ ëœ»í•©ë‹ˆë‹¤. 

ìƒí’ˆ $i$ì™€ ìƒí’ˆ $j$ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” ì•„ë˜ì™€ ê°™ì´ ì •ì˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
similarity(i,j)=\frac{\sum_{u \in U_{ij}} (r_{ui}-\bar{r_i}) \cdot (r_{uj}-\bar{r_j})}{\sqrt {\sum_{u \in U_{ij}} (r_{ui}-\bar{r_i})^2} \sqrt {\sum_{u \in U_{ij}} (r_{uj}-\bar{r_j})^2}}
$$

ë˜í•œ ì‚¬ìš©ì $u$ì™€ ì‚¬ìš©ì $v$ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ëŠ” ì•„ë˜ì™€ ê°™ì´ ì •ì˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
similarity(u,v)=\frac{\sum_{i \in I_{uv}} (r_{ui}- \bar {r_u}) \cdot (r_{vi}-\bar {r_v})}{\sqrt {\sum_{i \in I_{uv}} (r_{ui}- \bar {r_u})^2} \sqrt {\sum_{i \in I_{uv}} (r_{vi}-\bar {r_v})^2}}
$$

í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•  ë•Œ ì£¼ì˜í•´ì•¼ í•  ì ì´ ìˆìŠµë‹ˆë‹¤. ë°”ë¡œ ë‘ ë²¡í„°ê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ ê³„ì‚°í•˜ëŠ” ê³¼ì •ì—ì„œ ë¶„ëª¨ê°€ 0ì´ ë  ìˆ˜ ìˆë‹¤ëŠ” ì ì…ë‹ˆë‹¤. ë¶„ëª¨ê°€ 0ì´ë˜ëŠ” ê²½ìš° ë¶„ìë„ 0ì´ ë‚˜ì˜¤ëŠ”ë° ì´ëŸ¬í•œ ê²½ìš° ê²°ê³¼ê°€ 0ì´ ë˜ì–´ì•¼í•˜ì§€ë§Œ 0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ë‹¤ëŠ” ì—ëŸ¬ê°€ ë°œìƒí•©ë‹ˆë‹¤. ë”°ë¼ì„œ ë¶„ëª¨ê°€ 0ì´ë˜ì§€ ì•Šë„ë¡ ë§Œë“¤ë©° ë¶„ëª¨ê°€ 0ì´ ì•„ë‹ˆë”ë¼ë„ ê²°ê³¼ì— ì˜í–¥ì„ ì£¼ì§€ ì•ŠëŠ” ì ë‹¹íˆ ì‘ì€ ê°’ì„ ë”í•´ì£¼ì–´ì•¼ í•©ë‹ˆë‹¤. ì´ ë”í•˜ëŠ” ì‘ì€ ê°’ì„ ë³´í†µ **epsilon**ì´ë¼ê³  í•©ë‹ˆë‹¤. íŒŒì´ì¬ì—ì„œ **epsilon**ì€ ë³´í†µ 0.0001 ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

**get_pearson_correlation**

```python
# ë‘ ë²¡í„°ê°„ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” get_pearson_correlation í•¨ìˆ˜ë¥¼ êµ¬í˜„í•´ì£¼ì„¸ìš”.
# ë‘ ë²¡í„° u, v ì¤‘ì— í•œ ê°œë¼ë„ ê°’ì´ ë¹„ì–´ ìˆë‹¤ë©´ ê·¸ì— ëŒ€ì‘í•˜ëŠ” ê°’ë“¤ì€ ìœ ì‚¬ë„ë¥¼ êµ¬í•  ë•Œ ì œì™¸í•´ì£¼ì„¸ìš”.
# ë¶„ëª¨ê°€ 0ì´ ë˜ëŠ” ê²½ìš°ë¥¼ ì œì™¸í•˜ê¸° ìœ„í•´ ìµœì¢…ì ìœ¼ë¡œ ë¶„ëª¨ì— epsilonì„ ë”í•´ì£¼ì„¸ìš”.
def get_pearson_correlation(u, v, epsilon=0.0001):
    mask = np.isfinite(u) & np.isfinite(v)
    u = u[mask]
    v = v[mask]
    mother1 = np.sqrt(((u-u.mean())**2).sum())
    mother2 = np.sqrt(((v-v.mean())**2).sum())
    son = ((u-u.mean())*(v-v.mean())).sum()
    return son / ((mother1*mother2)+epsilon)
    # Write your code here!
```

**get_pearson_correlation_table**

```python
# ëª¨ë“  ì¡°í•©ì˜ ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” itertoolsì˜ productë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. 
from itertools import product

# ëª¨ë“  ì‚¬ìš©ìê°„ì˜ í”¼ì–´ìŠ¨ ìƒê´€ê³„ìˆ˜ë¥¼ êµ¬í•´ì£¼ëŠ” get_pearson_correlation_tableë¥¼ êµ¬í˜„í•´ì£¼ì„¸ìš”.
# ë¶„ëª¨ê°€ 0ì´ ë˜ëŠ” ê²½ìš°ë¥¼ ì œì™¸í•˜ê¸° ìœ„í•´ ìµœì¢…ì ìœ¼ë¡œ ë¶„ëª¨ì— epsilonì„ ë”í•´ì£¼ì„¸ìš”. epsilonì€ 0.0001ìœ¼ë¡œ ì„¤ì •í•˜ê² ìŠµë‹ˆë‹¤. 
def get_pearson_correlation_table(ratings, epsilon = 0.0001):
    corr_list = []
    person_com=list(product(ratings.index, repeat=2))
    for uname,vname in person_com:
        u, v = ratings.loc[uname], ratings.loc[vname]
        score = get_pearson_correlation(u,v,epsilon)
        corr = {
            'u':uname,
            'v':vname,
            'score':score
        }
        corr_list.append(corr)
    corr_df = pd.DataFrame(corr_list)
    corr_table = corr_df.pivot_table(values='score',index='u',columns='v')
    return corr_table
    
    # Wrie your code here!
```

### Average Correction

ìœ ì‚¬í•œ ìƒí’ˆ/ì‚¬ìš©ìì— ë” í° ê°€ì¤‘ì¹˜ë¥¼ ì£¼ì–´ íŠ¹ì • ìƒí’ˆì— ëŒ€í•œ ì‚¬ìš©ìì˜ ratings ê°’ì„ ì˜ˆì¸¡í•©ë‹ˆë‹¤. ìœ ì‚¬ë„ì— ë”°ë¼ ratingsì— ìœ ì‚¬ë„ê°€ ë†’ì€ kê°œì˜ ìƒí’ˆì— ë‚´ë ¤ì§„ ratingsì˜ í‰ê· /ìœ ì‚¬ë„ê°€ ë†’ì€ këª…ì˜ ì‚¬ìš©ìê°€ ë‚´ë¦° ratingsì˜ í‰ê· ì„ ì œì™¸í•œ ê°’ì˜ ê°€ì¤‘ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤. ì´ ê°’ì— ìƒí’ˆì— ë‚´ë¦° ratings ê°’/ì‚¬ìš©ìê°€ ë‚´ë¦° ratings ê°’ì˜ í‰ê· ì„ ìµœì¢…ì ìœ¼ë¡œ ë”í•˜ì—¬ ì˜ˆì¸¡í•©ë‹ˆë‹¤. ìƒí’ˆ/ì‚¬ìš©ìê°€ ë°›ê±°ë‚˜ í•˜ëŠ” ratings íŠ¹ì„±ì„ ë” ì˜ ë°˜ì˜í•©ë‹ˆë‹¤.

1. íŠ¹ì • ìœ ì‚¬ë„ ë©”íŠ¸ë¦­ì— ë”°ë¼ ì‚¬ìš©ì $u$ì™€ ìœ ì‚¬í•œ këª…(í˜¹ì€ ì´í•˜)ì˜ ì‚¬ìš©ìë¥¼ ê²°ì •í•©ë‹ˆë‹¤.
2. ìœ ì‚¬í•œ këª…ì˜ ì‚¬ìš©ìê°€ ìƒí’ˆ $i$ì— ë‚´ë¦° ratingsì— ê°ê° këª…ì˜ ì‚¬ìš©ìê°€ ë‚´ë¦° ratingsì˜ í‰ê· ì„ ëº€ ê°’ì— ëŒ€í•´ ìœ ì‚¬ë„ì— ë”°ë¥¸ ê°€ì¤‘ í‰ê· ì„ êµ¬í•©ë‹ˆë‹¤.
3. ì‚¬ìš©ì $u$ì´ ë‚´ë¦° ratings ê°’ì˜ í‰ê· ì„ ë”í•˜ì—¬ ê°’ì„ ë³´ì •í•´ì¤ë‹ˆë‹¤.
