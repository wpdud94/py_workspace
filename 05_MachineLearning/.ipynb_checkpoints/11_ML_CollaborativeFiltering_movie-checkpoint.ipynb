```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### 1. DataLoading and Data Preprocessing

```python
'''
ratings는 사용자별로 영화에 대한 평점을 매긴 데이터셋
평점은 최소 0.5~최대 5사이... 0.5단위로
'''
ratings = pd.read_csv('../data/ratings_1.csv', encoding='utf-8')
ratings.head()
# ratings.shape
# ratings['rating'].unique()
```

```python
movies = pd.read_csv('../data/movies.csv', encoding='utf-8')
movies.head()
# movies.shape
```

```python
# 필요한 컬럼만으로 다시 DataFrame 생성
ratings=ratings.iloc[:,:-1]
```

```python
# 사용자로 row, 영화정보를 컬럼으로 구성한 피벗 
ratings_matrix = pd.pivot_table(ratings, index='userId', columns='movieId', values='rating')
ratings_matrix
```

```python
'''
영화 title이 Id 대신 필요하다
title은 movies에 있다
merge 필요
'''
# 1. merge 후 pivot 화
rating_movies = pd.merge(ratings, movies, on='movieId')
ratings_matrix = pd.pivot_table(rating_movies, index='userId', columns='title', values='rating')

# 1-2. 결측치 0으로 처리
ratings_matrix=ratings_matrix.fillna(0)
ratings_matrix.head()
'''
NaN처리를 연구자의 몫
강사님은
1) NaN이 많으면 0
2) 적으면 mask
'''
```

### 영화와 영화들 간 유사도 산출 - Cosine Similarity
    사용자와 변환된 영화 평점 데이터 행렬 데이터셋을 이용해서 영화 간 유사도를 측정해보자
    코사인 유사도는 기준이 되는 다른 행을 비교해서 유사도를 특정하는 함수읻
    지금과 같은 행렬구조는 userId가 기준인 행 레벨 데이터이기 때문에 사용자 간 유사도가 만들어진다
    이것을 사용자가 아닌 영화 간 유사도를 만들려면 어떻게 해야 할까?
    
    참고로 코사인 유사도 라이브러리는 sklearn.metrics의 cosine_similarity()

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine_similarity?
```

```python
'''
유사도 진행 시 기준은 row

행과 열을 바꾸어야 하다

Transpose 해야 함
'''
```

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head()
```

```python
item_sim=cosine_similarity(ratings_matrix_T,ratings_matrix_T)
```

```python
item_sim = pd.DataFrame(item_sim, index=ratings_matrix_T.index, columns=ratings_matrix_T.index)
```

```python
item_sim
```
