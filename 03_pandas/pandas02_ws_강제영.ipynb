# 2. 엑셀파일 읽기 - 서울시 인구현황

```python
# 01. population_in_Seoul.xls 엑셀 파일을 읽으세요

```

```python
# 01. population_in_Seoul.xls 서울의 인구수에 대한 파일을 읽으세요. 
# 중복되는 header 파일을 제거하고 2번째부터 읽으세요 (header=2 이용)
# usecols = 'B, D, G, J, N' 옵션을 사용하여 해당 컬럼만 읽어 들이세요. 

Seoul_pop=pd.read_excel('../data/01. population_in_Seoul.xls',header=2,usecols='B, D, G, J, N')

Seoul_pop.head()
    
```

```python
# 각컬럼명을 '구별' '인구수', '한국인', '외국인', '고령자' 로 바꾸세요. 

Seoul_pop.rename(columns={
    '자치구':'구별',
    '계':'인구수',
    '계.1':'한국인',
    '계.2':'외국인',
    '65세이상고령자':'고령자'
}, inplace=True)

Seoul_pop.head()
```

# 3. CCTV 데이터 파악하기

```python
CCTV_Seoul.head()
```

```python
#'소계'로 오름차순 정렬하여 앞에서부터 5개만 보여주세요. 

CCTV_Seoul.sort_values(by='소계').head()
```

```python
#'소계'로 내림차순 정렬하여 앞에서부터 5개만 보여주세요. 

CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

```python
# '최근증가율' 컬럼에  (2016년도컬럼 + 2015년도컬럼 + 2014년도컬럼 )/ 2013년도 이전  * 100 으로 계산하여 채우세요.
# CCTV_Seoul['최근증가율']=np.nan
CCTV_Seoul['최근증가율']=(
    (CCTV_Seoul['2014년']+CCTV_Seoul['2015년']+CCTV_Seoul['2016년'])*100/CCTV_Seoul.iloc[:,2]
)
CCTV_Seoul

# 최근증가율 컬럼으로 내림차순 정렬하여 앞에서부터 5개만 보여주세요. 
CCTV_Seoul.sort_values(by='최근증가율',ascending=False).head()
```

# 4. 서울시 인구 데이터 파악하기

```python
Seoul_pop.head()
```

```python
# 위의 데이터중 첫번째 행을 지우세요. 합계정보는 필요없습니다. 

Seoul_pop.drop(0,inplace=True)
Seoul_pop.head()
```

```python
# '구별' 정보에서 unique() 함수를 사용하여 유니크한 값만 뽑으세요. 

Seoul_pop.iloc[:,0].unique()
```

```python
#pop_Seoul에 '구별' 정보가 null 인지 체크하세요. isnull() 함수 사용 

Seoul_pop[Seoul_pop.iloc[:,0].isnull()]
```

```python
# 위에서 찾은 26번째 행을 지우세요. 
# Seoul_pop.drop(Seoul_pop[Seoul_pop.iloc[:,0].isnull()].index, inplace=True)
Seoul_pop.dropna(inplace=True)
Seoul_pop[Seoul_pop.iloc[:,0].isnull()]
```

```python
# '고령자비율' 컬럼 = 고령자컬럼 / 인구수 컬럼 * 100 
# '외국인비율' 컬럼 = 외국인컬럼 / 인구수 컬럼 * 100

Seoul_pop['고령자비율']=(
    Seoul_pop['고령자']*100/Seoul_pop['인구수']
)
Seoul_pop['외국인비율']=(
    Seoul_pop['외국인']*100/Seoul_pop['인구수']
)
Seoul_pop.head()
```

```python
# '인구수'로 내림차순 정렬하여 앞에서부터 5개만 보여주세요. 


Seoul_pop.sort_values(by='인구수',ascending=False).head()
```

```python
# '외국인' 으로 내림차순 정렬하여 앞에서부터 5개만 보여주세요.

Seoul_pop.sort_values(by='외국인',ascending=False).head()
```

```python
# '외국인비율' 로 내림차순 정렬하여 5개만 보여주세요. 

Seoul_pop.sort_values(by='외국인비율',ascending=False).head()
```

```python
# '고령자' 로 내림차순 정렬하여 앞에서 부터 5개만 보여주세요. 


Seoul_pop.sort_values(by='고령자',ascending=False).head()
```

```python
#고령자비율 컬럼으로 내림차순 정렬하여 5개만 보여주세요. 

Seoul_pop.sort_values(by='고령자비율',ascending=False).head()
```
