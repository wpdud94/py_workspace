## 1. 법정동 코드 가공

```python
locCode = pd.read_excel('../data/법정동코드.xlsx')
locCode.head()
```

```python
locCode_seoul = locCode[locCode['시도명']=='서울특별시']
locCode_seoul = locCode_seoul[~locCode_seoul['읍면동명'].isnull()]
locCode_seoul = locCode_seoul[['법정동코드','시도명','시군구명','읍면동명','생성일자']]
locCode_seoul.reset_index(drop=True, inplace=True)
locCode_seoul= locCode_seoul.rename(columns={'읍면동명':'법정동'})
locCode_seoul.head()
```

## 2. 실거래가 데이터 가공

```python
apt = pd.read_csv('../data/2019년 서울 아파트 실거래가.csv', encoding="cp949",thousands=',')
apt
```

```python
# 거래금액 컬럼의 콤마(,) 제거 후 데이터 타입 변경..int64로
apt.head()
```

```python
# 전용면적(제곱미터)를 평으로 변환(나누기 3.306)하여 전용면적(평) 컬럼 생성 후 소숫점 둘쨰자리까지 표시합니다.
apt['전용면적(평)'] = round(apt['전용면적'] / 3.306,2)
apt.head()
```

```python
# 전용면적을 범주화 하기 위해 전용면적(평) 컬럼을 정수형으로 바꾼 새로운 컬럼을 생성합니다.
# 컬럼명 전용면적2(평)
apt['전용면적(평)2'] = [int(i) for i in apt['전용면적(평)']]
apt.head()
```

```python
# 전용면적을 범주화 합니다.
# 컬럼명은 전용면적(평)3
bins=np.arange(0,86,5)
bins
apt['전용면적(평)3'] = pd.cut(apt['전용면적(평)2'], bins)
apt.head()
```

```python
# 범주화된 전용면적의 고유값을 확인합니다.
apt['전용면적(평)3'].unique()
```

```python
locCode_seoul.head()
```

```python
apt.head()
```

```python
locCode_seoul['법정동코드'] = locCode_seoul['법정동코드'].apply(str)
locCode_seoul['지역코드'] = locCode_seoul['법정동코드'].str.slice(start=0, stop=5)
locCode_seoul['지역코드_법정동'] = locCode_seoul['지역코드']+locCode_seoul['법정동']
# locCode_seoul['법정동_읍면동'] = locCode_seoul['법정동코드'].str.slice(start=5, stop=8)
# locCode_seoul.sort_values(by='지역코드')
locCode_seoul.head()
```

```python
apt['지역코드'] = apt['지역코드'].apply(str)
apt.head()
```

```python
apt['법정동'] = apt['법정동'].str.slice(start=1)
apt.head()
```

```python
apt['지역코드_법정동'] = apt['지역코드']+apt['법정동']
apt
```

## 3. 데이타 프레임 병합

```python
#locCode.drop(['동리명','생성일자','말소일자'], axis=1, inplace=True)
merged = pd.merge(locCode_seoul,apt, on='지역코드_법정동')
merged
```

```python
# merged 컬럼 정라
merged.drop(columns='법정동_y',inplace=True)
merged.drop(columns='지역코드_y',inplace=True)
```

```python
# merged 컬럼 이름 정리
merged.rename(columns={'법정동_x':'법정동','지역코드_x':'지역코드'},inplace=True)
merged.head()
```

## 4. 데이타 분석


## 강제영_데이터 분석(시각화는 밑에)

```python
# 평당 금액
merged['평당 금액'] = round(merged['거래금액'] / merged['전용면적(평)'],2)
merged
```

```python
# 거래월 별 거래량
merged_exchageM_amount = merged.groupby('거래월')[['거래일자']].count()
merged_exchageM_amount.columns = ['거래량']
merged_exchageM_amount
```

```python
# 거래월 별 평균 매매가
merged_exchageM_price = merged.groupby('거래월').mean().loc[:,['거래금액','평당 금액']]
merged_exchageM_price.columns = ['월별 매매가','평당 금액']
merged_exchageM_price
```

```python
merged.head()
```

```python
# 지역별 매매가
merged_by_region = merged.pivot_table(values=['거래금액','평당 금액'],index=['시군구명'])
merged_by_region.sort_values(by='거래금액', ascending=False).head(5)
```

```python
# 지역별 매매가 상위 4개 지역
merged_by_region_top4=merged_by_region.sort_values(by='거래금액', ascending=False).head(4).index.values[:5]
merged_by_region_top4
```

```python
# 지역별 매매가 하위 4개 지역
merged_by_region_bottom4=merged_by_region.sort_values(by='거래금액').head(4).index.values[:5]
merged_by_region_bottom4
```

```python
# 지역별 월별 매매가
merged_by_region_month = merged.pivot_table(values=['거래금액','평당 금액'],index=['시군구명','거래월'])
# merged_by_month_region= merged_by_month_region.reset_index()
merged_by_region_month
```

```python
# 지역별 매매가 상위 4개 지역의 거래월별 지역별 매매가
merged_by_region_month_top4=merged_by_region_month.loc[merged_by_region_top4,:]
merged_by_region_month_top4_reset=merged_by_region_month_top4.reset_index()
merged_by_month_region_top4=merged_by_region_month_top4_reset.groupby(by=['거래월','시군구명']).mean()
merged_by_month_region_top4_reset= merged_by_month_region_top4.reset_index()
merged_by_month_region_top4_reset.head()
```

```python
# 지역별 매매가 하위 4개 지역의 거래월별 지역별 매매가
merged_by_region_month_bottom4=merged_by_region_month.loc[merged_by_region_bottom4,:]
merged_by_region_month_bottom4_reset=merged_by_region_month_bottom4.reset_index()
merged_by_month_region_bottom4=merged_by_region_month_bottom4_reset.groupby(by=['거래월','시군구명']).mean()
merged_by_region_month_bottom4_reset= merged_by_month_region_bottom4.reset_index()
merged_by_region_month_bottom4_reset.head()
```

```python
# 지역별 평균층수
import math
merged_floor_by_region = merged.groupby('시군구명')[['층']].mean()
merged_floor_by_region.sort_values(by='층', ascending=False)
merged_floor_price_by_region=(
pd.merge(merged_floor_by_region.reset_index(),merged_by_region.reset_index(),on='시군구명')
)

# 로그화
merged_floor_price_by_region['거래금액ln']=[math.log(i) for i in merged_floor_price_by_region['거래금액']]
merged_floor_price_by_region
merged['거래금액ln']=[math.log(i) for i in merged['거래금액']]
merged.head()
```
