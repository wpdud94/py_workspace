# 3. 데이터 가공 및 분석

```python
# 데이터 프레임 정보 조회
df.info()
```

```python
# 상표 컬럼의 고유값 조회
df['상표'].unique()
```

```python
# 지역 컬럼의 고유값 조회
df['지역'].unique()
```

```python
# 지역 컬럼의 값을 이용하여 시와 구를 분리
df['시']=[str(region.split(' ')[0]) for region in df['지역']]
df['구']=[str(region.split(' ')[1]) for region in df['지역']]
df.head()
```

```python
# 시 컬럼의 서울을 서울특별시로 변환
df['시'][(df['시']=='서울')] = '서울특별시'
df['시'].unique()
```

```python
# 기간 컬럼을 datetime 형식으로 변경
df['기간'] = pd.to_datetime(df['기간'],format='%Y%m%d')
df.head()
```

```python
# 기간 컬럼에서 dt 타입을 이용하여 년, 월, 일, 요일 컬럼 생성
# df['년']=df['기간'].dt.year
df['월']=df['기간'].dt.month
df['일']=df['기간'].dt.day
df['요일']=df['기간'].dt.day_name()

df
```

```python
# 휘발유 가격이 비싼 5개 판매가격을 출력하세요.
df.sort_values('휘발유',ascending=False).head()
```

```python
# 휘발유 가격이 저렴한 5개 판매가격을 출력하세요. (단 휘발유가 0원인 데이터는 제외)
df[df['휘발유']>0].sort_values('휘발유').head()
```

```python
# 상표별 평균 휘발유 가격을 계산하세요.
mean_gasolin = df.groupby(by='상표')[['휘발유']].mean()
mean_gasolin.columns = ['mean']
mean_gasolin
```

```python
# 기름 데이터 중 0 지우기
df_nozero = df[(df['휘발유']>0)&(df['고급휘발유']>0)&(df['경유']>0)]
df_nozero
```

```python
# 월별 기름 데이터 만들기
mean_oil_by_month=df_nozero.groupby('월').mean().iloc[:,:3]
mean_oil_by_month

# 월별 기름 변화량 데이터
mean_oil_by_month['고급휘발유L1']=mean_oil_by_month['고급휘발유'].shift(1)
mean_oil_by_month['휘발유L1']=mean_oil_by_month['휘발유'].shift(1)
mean_oil_by_month['경유L1']=mean_oil_by_month['경유'].shift(1)
mean_oil_by_month['고급휘발유증가율']=(mean_oil_by_month['고급휘발유']-mean_oil_by_month['고급휘발유L1'])/mean_oil_by_month['고급휘발유L1']*100
mean_oil_by_month['휘발유증가율']=(mean_oil_by_month['휘발유']-mean_oil_by_month['휘발유L1'])/mean_oil_by_month['휘발유L1']*100
mean_oil_by_month['경유증가율']=(mean_oil_by_month['경유']-mean_oil_by_month['경유L1'])/mean_oil_by_month['경유L1']*100

mean_oildelta_by_month = mean_oil_by_month.iloc[:,6:]
mean_oildelta_by_month
```

```python
# 월별 전체 기름 평균 가격 추이
mean_oil_by_month.iloc[:,:3].plot()
```
